```python
fibonacci = [1]
limit = 4000000
total = 0
isContinue = True

while isContinue:
    try:
        value = fibonacci[-1] + fibonacci[-3]
    except:
        value = fibonacci[-1] + 1
    
    fibonacci.append(value)
    print(fibonacci)
    
    if value < limit:
        if value % 2 == 0:
            total = total + value
            print("total " + str(total))
    else:
        isContinue = False
```
